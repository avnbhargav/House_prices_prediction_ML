```python
cols_to_drop = ['Id', 'MiscVal', 'Alley', 'PoolQC', 'Fence', 'MiscFeature']
df_train=df_train.drop(cols_to_drop,axis=1)
```

```python
df_train['LotFrontage']=df_train['LotFrontage'].fillna(df_train['LotFrontage'].median())
```

```python
df_train.shape
```

```python
bsmt_cols = ['BsmtQual', 'BsmtCond', 'BsmtExposure',
             'BsmtFinType1', 'BsmtFinType2']
for col in bsmt_cols:
    df_train[col] = df_train[col].fillna('NoBasement')

garage_cols = ['GarageType', 'GarageFinish', 'GarageQual', 'GarageCond']
for col in garage_cols:
    df_train[col] = df_train[col].fillna('NoGarage')

df_train['GarageYrBlt'] = df_train['GarageYrBlt'].fillna(0)

df_train['MasVnrType'] = df_train['MasVnrType'].fillna('None')
df_train['MasVnrArea'] = df_train['MasVnrArea'].fillna(0)

df_train['FireplaceQu'] = df_train['FireplaceQu'].fillna('None')

df_train['Electrical'] = df_train['Electrical'].fillna(df_train['Electrical'].mode()[0])

for col in df_train.select_dtypes(include=['object']).columns:
    df_train[col] = df_train[col].fillna('None')

for col in df_train.select_dtypes(include=['int64', 'float64']).columns:
    df_train[col] = df_train[col].fillna(df_train[col].median())
```

```python
qual_map={'None':0, 'Po':1, 'Fa':2, 'TA':3, 'Gd':4, 'Ex':5}
for col in ['ExterQual','ExterCond','BsmtQual','BsmtCond','HeatingQC','KitchenQual','FireplaceQu','GarageQual','GarageCond']:
    df_train[col] = df_train[col].map(qual_map).fillna(0)
```

```python
df_train['FireplaceQu'].unique()
```

```python
df_train['FireplaceQu'] = df_train['FireplaceQu'].replace('None', 0)
```

```python
def group_rare(series, thresh=0.01):
    freq=series.value_counts(normalize=True)
    rares=freq[freq<thresh].index
    return series.replace(rares,'other')
cat_cols = df_train.select_dtypes(include=['object']).columns.tolist()
for col in cat_cols:
    df_train[col] = group_rare(df_train[col], thresh=0.01)
```

```python
num_cols = df_train.select_dtypes(include=['int64','float64']).columns.tolist()
if 'SalePrice' in num_cols:
    num_cols.remove('SalePrice')
cat_cols = df_train.select_dtypes(include=['object']).columns.tolist()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.impute import SimpleImputer
```

```python
num_trans = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

cat_trans=Pipeline(steps=[
    ('imputer',SimpleImputer(strategy='constant',fill_value='None')),
    ('onehot',OneHotEncoder(handle_unknown='ignore',sparse=False))])

preprocessor = ColumnTransformer(transformers=[
    ('num', num_trans, num_cols),
    ('cat', cat_trans, cat_cols)
])
```

```python
from sklearn.model_selection import train_test_split
x=df_train.drop('SalePrice',axis=1)
y=df_train['SalePrice']
x_train,x_val,y_train,y_val=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.linear_model import Ridge

pipe_lr = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', Ridge(alpha=1.0))
])


pipe_rf = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', RandomForestRegressor(n_estimators=200, random_state=42, n_jobs=-1))
])

pipe_gb = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', GradientBoostingRegressor(n_estimators=200, learning_rate=0.05, random_state=42))
])
```

```python
from sklearn.model_selection import cross_val_score
import numpy as np

def cv_rmse(pipe):
    scores = cross_val_score(pipe, x_train, y_train,
                             scoring='neg_mean_squared_error',
                             cv=5, n_jobs=-1)
    rmse_scores = np.sqrt(-scores)
    return rmse_scores.mean(), rmse_scores.std()


for name,pipe in [('Linear',pipe_lr),('RF',pipe_rf),('GB',pipe_gb)]:
    mean_rmse,std_rmse=cv_rmse(pipe)
    print(f"{name}: RMSE={mean_rmse:.2f}±{std_rmse:.2f}")
```

```python


df_test_raw = df_test.copy()

qual_map = {'None':0, 'Po':1, 'Fa':2, 'TA':3, 'Gd':4, 'Ex':5}

qual_cols = [
    'ExterQual','ExterCond','BsmtQual','BsmtCond',
    'HeatingQC','KitchenQual','FireplaceQu','GarageQual','GarageCond'
]

for col in qual_cols:
    if col in df_test_raw.columns:
        df_test_raw[col] = df_test_raw[col].map(qual_map).fillna(0)

def group_rare(series, thresh=0.01):
    freq = series.value_counts(normalize=True)
    rares = freq[freq < thresh].index
    return series.replace(rares, 'other')

cat_cols_test = df_test_raw.select_dtypes(include=['object']).columns.tolist()

for col in cat_cols_test:
    df_test_raw[col] = group_rare(df_test_raw[col], thresh=0.01)

num_cols_test = df_test_raw.select_dtypes(include=['int64','float64']).columns.tolist()
cat_cols_test = df_test_raw.select_dtypes(include=['object']).columns.tolist()

print("Test preprocessing COMPLETE!")
print("Numeric cols:", len(num_cols_test))
print("Categorical cols:", len(cat_cols_test))
```
